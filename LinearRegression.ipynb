```python
#Importing libraries
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the Dataset
calfornia= datasets.fetch_california_housing()
print(calfornia)
```

```python
#Making dataframes
df_x=pd.DataFrame(calfornia.data,columns=calfornia.feature_names)
df_y=pd.DataFrame(calfornia.target)
```

```python
#Seeing first 5 x values
df_x.head()
```

```python
#Seeing first 5 y values
df_y.head()
```

```python
#Finding a few statistical values for the x values
df_x.describe()
```

```python
#Creating a model
reg=linear_model.LinearRegression()
```

```python
#Spliting values into test and train
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.33,random_state=42)
```

```python
reg.fit(x_train,y_train)
```

```python
#Predicting y values from the test values
y_pred=reg.predict(x_test)
```

```python
#Plotting a graph between actual and predicted prices
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
```

```python
#Calculating the mean squared error
print(mean_squared_error(y_pred,y_test))
```

```python
#Calculating the r2 score
score=r2_score(y_test,y_pred)
print(score)
```
