```python
#Importing libraries
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics
import numpy as np
import seaborn as sb
```

```python
#Loading the dataset
iris=datasets.load_iris()
print(iris)
```

```python
#Making dataframes
df_x=pd.DataFrame(iris.data,columns=iris.feature_names)
df_y=pd.DataFrame(iris.target)
```

```python
#Seeing first 5 x values
df_x.head()
```

```python
#Seeing first 5 y values
df_y.head()
```

```python
#Creating a model
iris_model=linear_model.LogisticRegression()
```

```python
#Spliting values into test and train
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.33,random_state=2)
```

```python
iris_model.fit(x_train,y_train)
```

```python
#Predicting y values from the test values
y_pred=iris_model.predict(x_test)
```

```python
#Creating a confusion matrix for the data
matrix=pd.crosstab(y_test.values.ravel(), y_pred, rownames=['Actual'], colnames=['Predicted'])
print(matrix)
```

```python
#Plotting a heat map for the confusion matrix
sb.heatmap(matrix,annot=True)
```

```python
#Calculating the model's accuracy
iris_model.score(x_test,y_test)
```
