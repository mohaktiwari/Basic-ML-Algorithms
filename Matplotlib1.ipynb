```python
import matplotlib.pyplot as plt
student1=[1,2,3,4,5]
marks1=[91,89,79,95,83]
plt.plot(student1,marks1,label="Student 1")

student2=[1,2,3,4,5]
marks2=[92,93,81,74,97]
plt.plot(student2,marks2,label="Student 2")

student3=[1,2,3,4,5]
marks3=[83,92,74,92,78]
plt.plot(student3,marks3,label="Student 3")

student4=[1,2,3,4,5]
marks4=[96,94,97,90,89]
plt.plot(student4,marks4,label="Student 4")

#Adding legends
plt.legend()

#Setting axes labels
plt.xlabel("Subjects")
plt.ylabel("Marks Obtained")

#Saving the plot as an image
plt.savefig("MyPlot.png")
plt.show()

#Setting axes limits
plt.xlim(1,3)
plt.ylim(80,90)
plt.plot(student1,marks1,student2,marks2,student3,marks3,student4,marks4)
plt.show()
```

```python
#Creating Subplots
import matplotlib.pyplot as plt
student1=[1,2,3,4,5]
marks1=[91,89,79,95,83]
plt.subplot(2,2,1)
plt.plot(student1,marks1,label="Student 1")
plt.legend()

marks2=[92,93,81,74,97]
plt.subplot(2,2,2)
plt.plot(student1,marks2,label="Student 2",color='orange')
plt.legend()

marks3=[83,92,74,92,78]
plt.subplot(2,2,3)
plt.plot(student1,marks3,label="Student 3",color='g')
plt.legend()

marks4=[96,94,97,90,89]
plt.subplot(2,2,4)
plt.plot(student1,marks4,label="Student 4",color='r')
plt.legend()

plt.show()
```
