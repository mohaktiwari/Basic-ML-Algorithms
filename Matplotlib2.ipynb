```python
#Line Plot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[78,90,67,83,99]
plt.plot(x,y)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Line Plot")
plt.show()
```

```python
#Area Plot
plt.fill_between(x,y)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Area Plot")
plt.show()
```

```python
#Scatter Plot
from sklearn import datasets
import numpy as np
iris=datasets.load_iris()
x=iris.data
y=iris.target
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#Simple Bar Plot
x=[1,2,3,4,5]
y=[78,90,67,83,99]
plt.bar(x,y,color='g')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Simple Bar Plot")
plt.show()
```

```python
#Grouped Bar plot
x=np.arange(1,6)
y1=[78,90,67,83,99]
y2=[77,91,67,76,91]
y3=[86,89,79,87,93]
y4=[78,93,82,82,91]
width=0.1
plt.bar(x-0.15,y1,width,color='g',label='Student 1')
plt.bar(x-0.05,y2,width,color='r',label='Student 2')
plt.bar(x+0.05,y3,width,color='b',label='Student 3')
plt.bar(x+0.15,y4,width,color='pink',label='Student 4')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Grouped Bar plot")
plt.legend()
plt.show()
```

```python
#Stacked Bar Plot
x=np.arange(1,6)
y1=np.array([78,90,67,83,99])
y2=np.array([77,91,67,76,91])
y3=np.array([86,89,79,87,93])
y4=np.array([78,93,82,82,91])
width=0.4
plt.bar(x,y1,width,color='g',label="Student 1")
plt.bar(x,y2,width,bottom=y1,color='r',label="Student 2")
plt.bar(x,y3,width,bottom=y1+y2,color='y',label="Student 3")
plt.bar(x,y3,width,bottom=y1+y2+y3,color='pink',label="Student 4")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Stacked Bar Plot")
plt.legend()
plt.show()
```

```python
#Histogram
import random
x=random.choices(range(0,101),k=1000)
plt.hist(x,bins=20,color='lavender',ec='black')
plt.title("Histogram")
plt.show()
```

```python
#Pie Plot
x=["Subject 1","Subject 2","Subject 3","Subject 4","Subject 5"]
y1=np.array([78,90,67,83,99])
plt.pie(y1,labels=x)
plt.title("Pie Plot")
plt.show()
```

```python
#Box Plot
y1=np.array([78,90,67,83,99])
y2=np.array([77,91,67,76,91])
y3=np.array([86,89,79,87,93])
y4=np.array([78,93,82,82,91])
box_plot=[y1,y2,y3,y4]
plt.boxplot(box_plot,patch_artist=True,labels=["Student 1","Student 2","Student 3","Student 4"])
plt.title("Box Plot")
plt.show()
```

```python
#Violin Plot
violin_plot=[y1,y2,y3,y4]
plt.violinplot(violin_plot,showmedians=True)
labels=["Student 1","Student 2","Student 3","Student 4"]
plt.xticks(np.arange(1,len(labels)+1),labels)
plt.title("Violin Plot")
plt.show()
```

```python
#Marginal Plot
import seaborn as sb
df=sb.load_dataset('iris')
sb.jointplot(x=df['sepal_length'],y=df['sepal_width'],kind='scatter')
plt.show()
```

```python
#Contour Plot
x=np.linspace(0,50,100)
y=np.linspace(0,50,100)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)+np.cos(Y)
plt.contour(X,Y,Z)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Contour Plot")
plt.show()
```

```python
#Heat Map
x=np.random.randint(low=1,high=11,size=(10,10))
print("The data to be plotted:",end="\n\n")
print(x,end="\n\n")
sb.heatmap(x)
plt.title("Heat Map")
plt.show()
```

```python
#3D Plot
ax=plt.axes(projection='3d')
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)+np.cos(Y)
ax.plot_surface(X,Y,Z,cmap="magma")
plt.title("3D Plot")
plt.show()
```
